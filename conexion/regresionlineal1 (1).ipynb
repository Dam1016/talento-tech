### Regresión Lineal de Horas vs. Nuevos Soles

```python
import pandas as pd

# Datos proporcionados por el usuario
data = {
    'X (HORAS)': [10, 15, 18, 20, 24, 30, 35, 40, 45, 48],
    'Y (NUEVOS SOLES)': [500, 800, 850, 950, 1100, 1300, 1550, 1700, 1900, 2100]
}

df = pd.DataFrame(data)
display(df.head())
```

Ahora, realizaremos la regresión lineal utilizando `scikit-learn`.

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Preparar los datos para el modelo
X = df['X (HORAS)'].values.reshape(-1, 1)  # Variable independiente (horas)
y = df['Y (NUEVOS SOLES)'].values        # Variable dependiente (nuevos soles)

# Crear y entrenar el modelo de regresión lineal
model = LinearRegression()
model.fit(X, y)

# Obtener el coeficiente (pendiente) y el intercepto
coefficient = model.coef_[0]
intercept = model.intercept_

print(f'Coeficiente (pendiente): {coefficient:.2f}')
print(f'Intercepto: {intercept:.2f}')
```

Finalmente, visualicemos los datos y la línea de regresión para entender la relación.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Predicciones del modelo para graficar la línea de regresión
y_pred = model.predict(X)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='X (HORAS)', y='Y (NUEVOS SOLES)', data=df, s=100, color='blue', label='Datos Originales')
plt.plot(X, y_pred, color='red', linewidth=2, label='Línea de Regresión')

plt.title('Regresión Lineal: Horas vs. Nuevos Soles')
plt.xlabel('X (HORAS)')
plt.ylabel('Y (NUEVOS SOLES)')
plt.grid(True)
plt.legend()
plt.show()
```
