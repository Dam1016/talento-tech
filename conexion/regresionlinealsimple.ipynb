```python
import pandas as pd
import statsmodels.api as sm

# Data provided by the user
data = {
    'X': [1, 2, 2, 3, 4, 5, 6, 7],
    'Y': [2, 3, 4, 4, 4, 6, 5, 7]
}
df = pd.DataFrame(data)

print("Original DataFrame:")
display(df.head())
```

Now, let's perform a simple linear regression. We will treat `X` as the independent variable and `Y` as the dependent variable.

```python
# Define dependent and independent variables
Y = df['Y']
X = df['X']

# Add a constant to the independent variable for the intercept term
X = sm.add_constant(X)

# Create and fit the OLS (Ordinary Least Squares) model
model = sm.OLS(Y, X)
results = model.fit()

# Print the regression summary
print(results.summary())
```

The regression summary above provides detailed statistical information about the model, including coefficients, R-squared, p-values, and more. If you'd like, I can also visualize the regression line on a scatter plot.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Create a scatter plot of the data points
plt.figure(figsize=(8, 6))
sns.scatterplot(x='X', y='Y', data=df, s=100)

# Plot the regression line using the coefficients from the OLS model
# The intercept is results.params[0] and the coefficient for X is results.params[1]
plt.plot(df['X'], results.predict(X), color='red', linewidth=2, label='Regression Line')

plt.title('Scatter Plot with Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

The plot above shows the data points and the fitted regression line. This visualization helps confirm the linear relationship and how well the model fits the data. Is there anything else you'd like to explore about this regression?

```python
# Calculate the product of X and Y and add it as a new column
df['X_times_Y'] = df['X'] * df['Y']

# Display the DataFrame with the new column
print("DataFrame with 'X_times_Y' column:")
display(df.head())
```

```python
# Calculate X squared and add it as a new column
df['X_squared'] = df['X'] ** 2

# Display the DataFrame with the new column
print("DataFrame with 'X_squared' column:")
display(df.head())
```

```python
# Calculate the sum of each numerical column
# We'll select only the numeric columns to sum
column_sums = df.select_dtypes(include='number').sum()

# Create a new DataFrame for the sum row
sum_df = pd.DataFrame([column_sums], index=['Sum'])

# Concatenate the original DataFrame with the sum row
df_with_sums = pd.concat([df, sum_df])

# Display the DataFrame with the new sum row
print("DataFrame with total sums row:")
display(df_with_sums)
```
