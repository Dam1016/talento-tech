| Header 1 | Header 2 | Header 3 |
|----------|----------|----------|

```python
import pandas as pd

data = {
    'y': [78.5, 74.3, 104.3, 87.6, 95.9, 109.2, 102.7, 72.5, 93.1, 115.9, 83.8, 113.3, 109.4],
    'x1': [7, 1, 11, 11, 7, 11, 3, 1, 2, 21, 1, 11, 10],
    'x2': [26, 29, 56, 31, 52, 55, 71, 31, 54, 47, 40, 66, 68],
    'x3': [6, 15, 8, 8, 6, 9, 17, 22, 18, 4, 23, 9, 8],
    'x4': [60, 52, 20, 47, 33, 22, 6, 44, 22, 26, 34, 12, 12]
}

df = pd.DataFrame(data)
display(df.head())
```

```python
import numpy as np

df['intercept'] = np.ones(len(df))

# Reorder columns to have 'intercept' as the first column, then 'y', then x1-x4
column_order = ['intercept', 'y'] + [col for col in df.columns if col.startswith('x')]
df_matrix = df[column_order]

display(df_matrix)
```

```python
df_matrix_transpose = df_matrix.T
display(df_matrix_transpose)
```

```python
y_vector = df['y'].values.reshape(-1, 1)

# Display the shape and content of y_vector to confirm it's a column vector
print(f"Shape of y_vector: {y_vector.shape}")
print("Content of y_vector:")
print(y_vector)
```

```python
X = df_matrix[['intercept', 'x1', 'x2', 'x3', 'x4']].values
X_transpose = X.T

# Perform matrix multiplication: X_transpose * X
matrix_product = np.dot(X_transpose, X)

print("Shape of X:", X.shape)
print("Shape of X_transpose:", X_transpose.shape)
print("Shape of matrix_product (X_transpose * X):")
print(matrix_product.shape)
print("\nResult of (X_transpose * X):")
print(matrix_product)
```

```python
try:
    matrix_product_inverse = np.linalg.inv(matrix_product)
    print("Inverse of the matrix product:")
    print(matrix_product_inverse)
except np.linalg.LinAlgError:
    print("Error: The matrix is singular and cannot be inverted.")
```

```python
# Calculate (X_transpose * y)
X_transpose_y = np.dot(X_transpose, y_vector)

# Calculate betas: (X_transpose * X)_inverse * (X_transpose * y)
betas = np.dot(matrix_product_inverse, X_transpose_y)

print("Beta Coefficients:")
print(betas)
```
