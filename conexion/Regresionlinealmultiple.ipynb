```python
import pandas as pd
from io import StringIO
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
# Data provided by the user
data = """Y	X1	X2	X3	X4
78,5	7	26	6	60
74,3	1	29	15	52
104,3	11	56	8	20
87,6	11	31	8	47
95,9	7	52	6	33
109,2	11	55	9	22
102,7	3	71	17	6
72,5	1	31	22	44
93,1	2	54	18	22
115,9	21	47	4	26
83,8	1	40	23	34
113,3	11	66	9	12
109,4	10	68	8	12"""

# Read the data into a pandas DataFrame, replacing comma with dot for decimals
df = pd.read_csv(StringIO(data), sep='\t').replace(',', '.', regex=True)

# Convert all columns to numeric type
for col in df.columns:
    df[col] = pd.to_numeric(df[col])

display(df.head())
```

```python
# Define dependent variable (Y) and independent variables (X)
X = df[['X1', 'X2', 'X3', 'X4']]
Y = df['Y']

# Initialize and fit the Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Display the results
print(f"Coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")
```

The regression equation is of the form:
$Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \beta_3 X_3 + \beta_4 X_4$

Where:
- $\beta_0$ is the intercept.
- $\beta_i$ are the coefficients for each independent variable $X_i$.

From the results:
- The intercept ($\beta_0$) is approximately `{:.2f}`.
- The coefficients ($\beta_1, \beta_2, \beta_3, \beta_4$) are approximately `{:.2f}, {:.2f}, {:.2f}, {:.2f}` respectively.

This means for every one-unit increase in $X_1$, $Y$ increases by `{:.2f}` (holding other variables constant), and similarly for other variables. The intercept is the predicted value of $Y$ when all $X$ variables are zero.
