```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Leer archivo csv
df = pd.read_csv("ventas_tienda_online.csv")
df


```

```python
# mostrar las primeras filas del dataFrame
print(df.head())
```

```python
# Mostrar las ultimas filas del DataFrame
print(df.tail)
```

```python
# Informacion del Dataframe
print(df.describe)
```

```python
# Convertir la columna "fecha" a tipo datatime
df["fecha"] = pd.to_datetime(df['Fecha'])
print(df)
```

```python
# Agregar una columna de mes
df['mes'] = df['fecha'].dt.to_period('M')
print(df)
```

```python
# Calcular las ventas totales por mes
ventas_totales = df.groupby('mes')['Ingresos'].sum()
print(ventas_totales)
```

```python
# Calcular la cantidad total vendida por producto
cantidad_vendida = df.groupby('Producto')['Cantidad'].sum()
print(cantidad_vendida)
```

```python
# Calcular los ingresos totales por categoria
ingresos_totales = df.groupby('Categoría')['Ingresos'].sum()
print(ingresos_totales)
```

```python
# Calcular las ventas mensuales
ventas_mensuales = df.groupby(['mes', 'Producto'])['Cantidad'].sum()
df.to_csv('ventas_anuales.csv', index=False)
print(ventas_mensuales)
```

```python
# Grafico de lineas de ventas mensuales
ventas_mensuales.plot(kind="line", marker="o")
plt.title("ventas Mensuales")
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.title('Ventas Mensuales')
plt.show()
```

```python
# Grafico de barras de ingresos por categoria
ingresos_totales.plot(kind="bar", color="skyblue")
plt.title("Ingresos por Categoria")
plt.xlabel('Categoria')
plt.ylabel('Ingresos')
plt.xticks(rotation=45)
plt.show()
```
